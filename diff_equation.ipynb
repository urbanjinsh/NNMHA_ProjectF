```python
from typing import List
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

1-D ODE

$$
y' = x+3 \\
y(0) = 3
$$

```python
def diff_equation(y,x):
    return(np.array([x+3]))

# dy= lambda y,x:x+3

x = np.linspace(0, 10, 100)  # There are 100 point which start from 0 to 10
result = odeint(diff_equation, 3, x)  
```

```python
plt.plot(x, result[:, 0], label='y')

plt.legend()
plt.grid()  
plt.show() 
```

2-D ODE

$$
y_1' = y_2 \\
y_2' = -2 * y_1 - 2 * y_2\\
y_1(0) = 0, y_2(0) = 1
$$

```python
def diff_equation(y_list: List[float], x):
    y1, y2= y_list
    return np.array([y2, -2 * y1 - 2 * y2])  
# Format of differential equations. Left side is the dy/dx
 
x = np.linspace(0, 10, 100)  # There are 100 point which start from 0 to 10
result = odeint(diff_equation, [0,1], x) 
```

```python
plt.plot(x, result[:, 0], label='y1')
plt.plot(x, result[:, 1], label='y2')

plt.legend()
plt.grid()  
plt.show() 
```

3-D ODE

$$
y_1' = y_2 * y_3 \\
y_2' = -y_1 * y_3 \\
y_3' = -0.5y_1 * y_2 \\
y_1(0) = 0, y_2(0) = 1, y_3(0) = 2
$$

```python
def diff_equation(y_list: List[float], x: float):
    y1, y2, y3 = y_list
    return np.array([y2 * y3, -y1 * y3, -0.51 * y1 * y2])  
# Format of differential equations. Left side is the dy/dx
 
x = np.linspace(0, 10, 100)  # There are 100 point which start from 0 to 10
result = odeint(diff_equation, [0, 1, 2], x)  
# The first para is the differential equation
# The second para is the initial value of y
# The third para is the x value
```

```python
# plot the data
plt.plot(x, result[:, 0], label='y1')
plt.plot(x, result[:, 1], label='y2')
plt.plot(x, result[:, 2], label='y3')

plt.legend()
plt.grid()  
plt.show() 
```
